```python
from summa import summarizer
```

```python
#read the combined data
cmb_data = pd.read_csv(r'C:\Users\Sithalrao\Downloads\combined_data_11.csv')
```

```python
cmb_data['reviewText'] = cmb_data['reviewText'].astype(str)
```

```python
# Extract the first 1000 "rating-1.0
low_rating_reviews = [review for review, rating in zip(cmb_data['reviewText'], cmb_data['overall']) if rating == 1.0][:1000]
low_rating_reviews
```

```python
# Summarize to 1% of the words
low_rating_summary = summarizer.summarize(' '.join(low_rating_reviews), ratio=0.01)
print(low_rating_summary)
```
