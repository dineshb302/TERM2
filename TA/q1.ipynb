```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#read the combined data
cmb_data = pd.read_csv(r'C:\Users\Sithalrao\Downloads\combined_data_11.csv')
```

```python
#Extract review overall and reveiw text
cmb_data=cmb_data[['reviewText']]
cmb_data.count()
```

```python
#first and last 10000 review texts from amazon food 
first_10K=cmb_data.head(10000)
first_10K
```

```python
#last 10000 review texts from electronics
tail_10K=cmb_data.tail(10000)
tail_10K
```

```python
total_10K=pd.concat([first_10K,tail_10K], ignore_index=True)
total_10K
```

```python
#Q1 lowercasing and removing punctuation

total_10K['reviewText_processed']=total_10K['reviewText'].str.lower()
total_10K
```

```python
import string
```

```python
#remove punctuation
total_10K['reviewText_processed']=total_10K['reviewText_processed'].str.translate(str.maketrans('', '', string.punctuation))
total_10K
```

```python
total_10K['reviewText_processed']=total_10K['reviewText_processed'].astype(str)
```

# Compute IDF of all words in these reviews. Report the top 30 words and bottom 30 words, based on IDF, with their IDF scores

```python
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(total_10K['reviewText_processed'])
idf = vectorizer.idf_
idf
```

```python
terms_idf = dict(zip(vectorizer.get_feature_names_out(), idf))
sorted_words_by_idf = sorted(terms_idf.items(), key=lambda x: x[1])
column_names = ['word', 'IDF_score']
sorted_words_by_idf=pd.DataFrame(sorted_words_by_idf,columns=column_names)
sorted_words_by_idf=sorted_words_by_idf.reset_index(drop=True)

top_30_words = sorted_words_by_idf[-30:][::-1]  # Top 30 words
bottom_30_words = sorted_words_by_idf[:30]
top_30_words
```
